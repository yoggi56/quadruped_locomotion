```python
from locomotion_control_level.cpg_gait_generator import CPGGaitGenerator
from math import *
import matplotlib.pyplot as plt
```

```python
WALK = 0
TROT = 1
PACE = 2
GALLOP = 3

dt = 1/300
```

Walking gait

```python
cpg = CPGGaitGenerator()
cpg.set_static_params(alpha=100)
cpg.init_integrator(dt=dt)
```

```python

gait_dir = 0
gait_rot_dir = 0
w_sw = 2*pi
phi = WALK

phase_lst = [[],[],[],[]]
t = []
it = 0

for i in range(2500):
    cpg.set_gait_params(w_sw=w_sw, 
                            phi=phi, 
                            Hg=0.05, 
                            Ls=0.05, 
                            dir=gait_dir, 
                            rot_dir=gait_rot_dir, 
                            rot_r=0.5)
    ef_ref_pos = cpg.step()
    phase1, phase2, phase3, phase4 = cpg.get_phase()
    phase_lst[0].append(phase1)
    phase_lst[1].append(phase2)
    phase_lst[2].append(phase3)
    phase_lst[3].append(phase4)
    t.append(it)
    it += dt
    
```

Plots for phase outputs

```python
fig = plt.figure(figsize=(7.5,7.5),
            facecolor='white')

plt.subplot(4, 1, 1)

plt.title("Walk")
plt.grid()
plt.plot(t, phase_lst[0])
plt.ylabel("$\phi_1$")

plt.subplot(4, 1, 2)
plt.grid()
plt.plot(t, phase_lst[1])
plt.ylabel("$\phi_2$")

plt.subplot(4, 1, 3)
plt.grid()
plt.plot(t, phase_lst[2])
plt.ylabel("$\phi_3$")

plt.subplot(4, 1, 4)
plt.grid()
plt.plot(t, phase_lst[3])
plt.ylabel("$\phi_4$")
plt.xlabel("t [sec]")
```

Gait transition

```python
WALK = 0
TROT = 1
PACE = 2
GALLOP = 3

dt = 1/300

cpg = CPGGaitGenerator()
cpg.set_static_params(alpha=100)
cpg.init_integrator(dt=dt)

gait_dir = 0
gait_rot_dir = 0
w_sw = 2*pi
phi = WALK

phase_lst = [[],[],[],[]]
t = []
it = 0

while it < 5:
    if it < 2.5:
        phi = WALK
    else:
        phi = TROT
    
    cpg.set_gait_params(w_sw=w_sw, 
                            phi=phi, 
                            Hg=0.05, 
                            Ls=0.05, 
                            dir=gait_dir, 
                            rot_dir=gait_rot_dir, 
                            rot_r=0.5)
    ef_ref_pos = cpg.step()
    phase1, phase2, phase3, phase4 = cpg.get_phase()
    phase_lst[0].append(phase1)
    phase_lst[1].append(phase2)
    phase_lst[2].append(phase3)
    phase_lst[3].append(phase4)
    t.append(it)
    it += dt
```

```python
# vertical line
trans_point_t = [2.5, 2.5]
trans_point_phi = [-4, 4]

trans_endpoint_t = [3.6, 3.6]
trans_endpoint_phi = [-4, 4]

fig = plt.figure(figsize=(15,5),
            facecolor='white')

plt.title("Gait transition")
plt.grid()
plt.plot(t, phase_lst[0], label="RF")
plt.plot(t, phase_lst[1], label="LF")
plt.plot(t, phase_lst[2], label="RH")
plt.plot(t, phase_lst[3], label="LH")
plt.plot(trans_point_t, trans_point_phi, 'k--')
plt.plot(trans_endpoint_t, trans_endpoint_phi, 'k--')
plt.ylabel("$\phi_i$")
plt.xlabel("t [sec]")
plt.legend(title='Legs')
```
